```python
import os
import pandas as pd
import numpy as np

from joblib import load

import geopandas as gpd
import model_prep_and_evals as mpe # custom module
```

```python
rfc = load('window_lidar_spectral_rfc.joblib') 
```

```python
pts = pd.read_csv(os.path.join(os.getcwd(), 
                               'validation_results_spectral_2020', 
                               'validation_results_spectral_2020.csv')).loc[:,'r':'avg_lidar']
pts.head()
```

```python
preds = rfc.predict(pts.to_numpy())
```

```python
pts_ref_class = pd.read_csv(os.path.join(os.getcwd(), 'validation_results_spectral_2020', 'validation_results_spectral_2020.csv')).loc[:,'ref_class']
```

```python
def to_0_and_1(multi_class):
    simple = []
    for x in multi_class:
        if x > 1:
            simple.append(0)
        else:
            simple.append(x)
    return simple

ground = to_0_and_1(pts_ref_class)
```

```python
mpe.print_accuracy_info(np.array(ground), preds)
```

```python
pts_map_class = pd.read_csv(os.path.join(os.getcwd(), 'validation_results_spectral_2020', 'validation_results_spectral_2020.csv')).loc[:,'map_class']
map_class = to_0_and_1(pts_map_class)
```

```python
mpe.print_accuracy_info(np.array(ground), np.array(map_class))
```
