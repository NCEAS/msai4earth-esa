## Example 0: creating a 2-dimensional xarray.DataArray

```python
# values of a single variable at each point of the coords (it's a numpy.ndarray)
sdp_data = np.array([[1,2],[0,1]])  

# names of the data dimensions (tuple of strings)
dims = ('lat','lon')        

# coordinates (tick labels) to use for indexing along each dimension (it's a dictionary)
coords = {'lat' : [10, 20],
          'lon' : [30, 40]}   

# attributes (metadata) of the data array (it's a dictionary)
attrs = { 'creator_name': 'Carmen GG', 
          'title' : 'single day precipitation',
          'units' : 'ml / pixel'}
```

```python
sdp = xr.DataArray(data = sdp_data, 
                    dims = dims,
                    coords = coords,
                    attrs = attrs)
sdp
```

```python
# we can also update the attributes (metadata) after creating the set
sdp.attrs['description'] = 'this is a small example to understand how to create an xarray datarray'

# each of the coordinates is also an xarray.DataArra
# so we can add attributes to coordinates 
sdp.lat.attrs['standard_name']= 'latitude'
sdp.lat.attrs['units'] = 'degrees_north'

sdp.lon.attrs['standard_name']= 'longitude'
sdp.lon.attrs['units'] = 'degrees_east'
sdp
```

## Example 2: adding one more dimension

```python
# suppose we take measurements for precipitation on 3 days
mdp_data = np.array([[[1,2],[0,1]],
                      [[2,2],[2,2]],
                      [[0,0],[0,0]]
                     ])
mdp_data[0]
```

```python
# dimensions
dims = ('day', 'lat', 'lon')

# coordinates (tick labels) to use for indexing along each dimension (dictionary)
coords = {'day' : [1,2,3],
          'lat' : [10, 20],
          'lon' : [30, 40]} 

mdp = xr.DataArray(data = mdp_data,
                              dims = dims,
                              coords = coords, 
                              attrs = sdp.attrs)

mdp.lat.attrs = sdp.lat.attrs
mdp.lon.attrs = sdp.lon.attrs
mdp.day.attrs = {'standard_name':'day of collection'}
mdp
```

```python
# update title
mdp.attrs['title'] = 'precipitation data on days 1, 2, and 3'
mdp.attrs['title']
```
